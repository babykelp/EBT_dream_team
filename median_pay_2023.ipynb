```python
import pandas as pd
import numpy as np
```

```python
tab_23 = pd.read_excel('tab_23.xls', skiprows = 4)
tab_23 = tab_23.set_index('Unnamed: 0')
```

```python
median_zp = pd.DataFrame(index = np.arange(0,96))
median_zp['object_name']= tab_23.index
median_zp['median_pay']= tab_23.index
for i, idx in enumerate(tab_23.index):
    cumulate_pay = 0
    for j, clm in enumerate(tab_23.columns):
        cumulate_pay += float(tab_23.iloc[i,j+1])
        if cumulate_pay >= 50:
          numbers = clm.replace("от", "").replace("до", "").split()
          first_number = float(numbers[0].replace(",", "."))
          try:
            second_number = float(numbers[1].replace(",", "."))

            # Вычисление среднего значения
            average_value = (first_number + second_number) / 2
          except:
            average_value = first_number
          median_zp.loc[i,'median_pay'] = average_value
          break
# median_zp.to_csv('median_pay_23.csv')
median_zp
```
